```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assurez-vous que df_labled existe avec les colonnes 'comment' et 'score'.

# Tokenize les commentaires
df_labled['tokens'] = df_labled['comment'].str.split()

# Crée une table de contingence des fréquences des mots par classe
word_class_freq = pd.DataFrame([(word, row['score'])
                                for _, row in df_labled.iterrows()
                                for word in row['tokens']],
                               columns=['word', 'class'])
contingency_table = pd.crosstab(word_class_freq['word'], word_class_freq['class'])

# Calculer la fréquence totale de chaque mot et sélectionner les 50 mots les moins fréquents
total_frequencies = contingency_table.sum(axis=1)
least_frequent_words = total_frequencies.nsmallest(50).index

# Filtrer la table de contingence pour inclure seulement les 50 mots les moins fréquents
filtered_contingency_table = contingency_table.loc[least_frequent_words]

# Trace la table de contingence filtrée sous forme de heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(filtered_contingency_table, annot=True, cmap='viridis')
plt.xlabel('Class')
plt.ylabel('Word')
plt.title('Word Frequencies by Class for the 50 Least Frequent Words')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder


# Convertir les commentaires en une matrice de termes
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_labled['comment'])

# Encoder les classes comme des valeurs numériques
label_encoder = LabelEncoder()
encoded_classes = label_encoder.fit_transform(df_labled['score'])

# Effectuer le test du Chi-deux
chi2score = chi2(X, encoded_classes)

# Associer les termes avec leurs scores de chi-deux
scores_with_terms = list(zip(vectorizer.get_feature_names_out(), chi2score[0]))

# Trier par scores de chi-deux en ordre croissant
sorted_scores_with_terms = sorted(scores_with_terms, key=lambda x: x[1])

# Sélectionner les 600 termes avec les scores les plus bas
lowest_600_scores_with_terms = sorted_scores_with_terms[:600]

# Créer un DataFrame à partir des 600 termes les moins corrélés
lowest_600_df = pd.DataFrame(lowest_600_scores_with_terms, columns=['Word', 'Chi2 Score'])

# Tracer un barplot horizontal des scores
plt.figure(figsize=(10, 8))
sns.barplot(x='Chi2 Score', y='Word', data=lowest_600_df, palette='viridis')
plt.xlabel('Chi2 Score')
plt.ylabel('Word')
plt.title('600 Words with Lowest Chi2 Scores')
plt.show()
```

```python
# Afficher les 50 termes avec les scores les plus bas
for word, chi2_score in lowest_600_scores_with_terms:
    print(f"{word}: {chi2_score}")
```
